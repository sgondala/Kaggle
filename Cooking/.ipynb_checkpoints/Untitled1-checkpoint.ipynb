```python
import json
```

```python
train = json.load(open("train.json"))
```

```python
cuisineMap = {} #Contains cuisine sum, not prob
cuisineIndex = {}
cuisineCount = 0
```

```python
for i in train:
    cuisine = i['cuisine']
    if cuisine in cuisineMap:
        cuisineMap[cuisine] += 1
    else:
        cuisineMap[cuisine] = 1
        cuisineIndex[cuisine] = cuisineCount
        cuisineCount += 1
```

```python
cuisineSum = 0
for i in cuisineMap:
    cuisineSum += cuisineMap[i]
```

```python
len(train), cuisineSum
```

```python
ingredientMap = {} #Contains sum, not prob
ingredientSum = 0
```

```python
for i in train:
    for j in i['ingredients']:
        ingredientSum += 1
        if j in ingredientMap:
            ingredientMap[j] += 1
        else:
            ingredientMap[j] = 1
```

```python
ingredientSum
```

```python
igc = {}
for i in range(20):
    igc[i] = {}
```

```python
igc
```

```python
for i in train:
    cuisine = i['cuisine']
    index = cuisineIndex[cuisine]
    ingredientList = i['ingredients']
    for ingredient in ingredientList:
        if ingredient in igc[index]:
            igc[index][ingredient] += 1.0
        else:
            igc[index][ingredient] = 1.0
```

```python
igc[19]
```

```python
igcCount = {}
for i in range(20):
    igcCount[i] = 0.0
```

```python
for i in range(20):
    for ingredient in igc[i]:
        igcCount[i] += igc[i][ingredient]
```

```python
tempSum = 0
for i in range(20):
    for j in igc[i]:
        tempSum += igc[i][j]
tempSum
```

```python
for i in range(20):
    for j in igc[i]:
        igc[i][j] /= igcCount[i]
```
